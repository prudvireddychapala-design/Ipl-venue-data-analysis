```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("matches.csv")
df.head()
df.info()
df.describe()
team_wins = df['winner'].value_counts()

plt.figure()
team_wins.head(10).plot(kind='bar')
plt.title("Top 10 Most Successful IPL Teams")
plt.xlabel("Team")
plt.ylabel("Wins")
plt.show()
toss_match_win = df[df['toss_winner'] == df['winner']]
percentage = (len(toss_match_win) / len(df)) * 100
percentage
venue_matches = df['venue'].value_counts().head(10)

plt.figure()
venue_matches.plot(kind='bar')
plt.title("Top 10 IPL Venues by Matches Played")
plt.xlabel("Venue")
plt.ylabel("Matches")
plt.show()
df['result'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Match Result Types")
plt.ylabel("")
plt.show()




```
